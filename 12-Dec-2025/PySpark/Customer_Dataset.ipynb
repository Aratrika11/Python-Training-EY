```python
#Never assign .show() to a variable
#Show the first 10 customers.
dfr.show(10)
```

```python
#List all unique cities.
dfr.select("city").distinct().show()
```

```python
#Display only customer_id, name, and annual_spend
dfr.select("customer_id", "name", "annual_spend").show()
```

```python
#Filter customers who spend more than 60000 annually
dfr.filter(dfr.annual_spend > 60000).show()
```

```python
#Show all customers from Delhi who are younger than 30
dfr.filter((dfr.city == "Delhi") & (dfr.age < 30)).show()
```

```python
#Create a new column spend_lakh = annual_spend / 100000
dfr = dfr.withColumn("spend_lakh", dfr.annual_spend / 100000)
dfr.show()
```

```python
"""Create customer_type column
spend > 70000 → Premium
else → Standard"""

df2 = dfr.withColumn(
    "customer_type",
    F.when(dfr.annual_spend > 70000, "Premium")
     .otherwise("Standard")
)
df2.show()
```

```python
#Show customers whose name starts with letter A
dfr.filter(dfr.name.startswith("A")).show()
```

```python
#Filter customers where category is Clothing or Electronics
dfr.filter(dfr.category.isin("Clothing", "Electronics")).show()
```

```python
#Convert city name to uppercase
df3 = dfr.withColumn("city_upper", F.upper(dfr.city))
df3.show()
```

```python
#Remove the category column

df2 = df2.drop("category")
df2.show()
```

```python
#Sort customers by age in descending order

dfr.orderBy(dfr.age.desc()).show()
```

```python
#Find customers younger than the average age
avg_age = dfr.select(F.avg("age")).collect()[0][0]

dfr.filter(dfr.age < avg_age).show()
```

```python
#Display top 5 highest spending customers

dfr.orderBy(dfr.annual_spend.desc()).show(5)
```

```python
#Create a new DataFrame with only customers from Mumbai

df_mumbai = dfr.filter(dfr.city == "Mumbai")
df_mumbai.show()
```

```python
#Bonus : Extract first letter of each name

df4 = dfr.withColumn("name_initial", F.substring(dfr.name, 1, 1))
dfr.show()
```

```python
#Bonus: Mask the customer_id (Example: C001 → 01)

df5 = dfr.withColumn(
    "masked_customer_id",
    F.concat(F.lit("**"), F.substring(dfr.customer_id, -2, 2))
)
df5.show()
```

```python
#Bonus : Create boolean column is_senior where age > 35

df6 = dfr.withColumn("is_senior", dfr.age > 35)
dfr.show()
```

```python
#Bonus : Count number of customers per city

dfr.groupBy("city").count().show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("Columnar Operations").getOrCreate()
```

```python
data = [
    ("U001","Delhi",450),
    ("U002","Mumbai",620),
    ("U003","Bangalore",300),
    ("U004","Delhi",700)
]
```

```python
columns = ["user_id","city","amount"]
df = spark.createDataFrame(data,columns)
df.show()
```

```python
df = df.withColumn("amount_with_tax",col("amount")*1.18)
df.show()
```

```python
df = df.withColumnRenamed("amount_with_tax","total_amount")
df.show()
```

```python
df =df.replace("Delhi","New Delhi",subset=["city"])
df.show()
```

```python
from pyspark.sql.functions import when
df = df.withColumn("amount_category",when(col("total_amount")>500,"High")
.otherwise("Low"))
df.show()
```
